```python
import pandas as pd
import sqlite3
```

```python
# To setup a connection on SQLite
def new_connection(dbfile):
    conn=sqlite3.connect(dbfile)
    return conn
```

```python
# To create a new table
def create_table(database, create_new_table):
    connection = sqlite3.connect(database)
    cursor=connection.cursor()
    cursor.execute(create_new_table)
    connection.commit()
    connection.close()
```

```python
def type2(col_type):
    if col_type=="int64":
        type1="INTEGER(64)"
    elif col_type=="float64":
        type1="FLOAT(64)"
    else:
        type1="VARCHAR(64)"
    return type1

def generate_string(tablename, df, primarykey):
    
    columnnames=list(df.columns.values)
    
    sql_string="CREATE TABLE IF NOT EXISTS"+ " "+ tablename + "("
    
    for i in range(len(columnnames)):
        
        if i==len(columnnames)-1:
            col=columnnames[i]
            col_type=df[columnnames[i]].dtype
        
            if col==primarykey:
                coltype=type2(col_type)
                sql_string = sql_string + col + " " + coltype + " " + "PRIMARY KEY" + " " + ")" + ";" 
            else:
                coltype=type2(col_type)
                sql_string= sql_string + col + " " + coltype + " " + ")" + ";"
        
        else:
            
            col=columnnames[i]
            col_type=df[columnnames[i]].dtype
        
            if col==primarykey:
                coltype=type2(col_type)
                sql_string = sql_string + col + " " + coltype + " " + "PRIMARY KEY" + ","
            else:
                coltype=type2(col_type)
                sql_string= sql_string + col + " " + coltype + ","
    
    return (sql_string)
```

```python
bookings_table=generate_string("bookings1", data_raw, "orig_index" )
```

```python
create_table("mldata2.db",bookings_table)
```

```python
# Pushing data into the database
def push_dataframe(df,database, tablename):
    connection = sqlite3.connect(database)
    cursor = connection.cursor()
    data_raw.to_sql(tablename,connection, if_exists="replace")
    connection.commit()
    connection.close()
```

```python
push_dataframe(data_raw,"mldata2.db","bookings1")
```

```python
# Query data from database

def query_data(database, query):
    connection = sqlite3.connect(database)
    cursor = connection.cursor()
    
    data=pd.read_sql(query,connection)
    connection.commit()
    connection.close()
    return data
```
